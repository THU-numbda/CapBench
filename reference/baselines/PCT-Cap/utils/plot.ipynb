```python
# raw dataset capacitance distribution detail
import matplotlib.pyplot as plt
import matplotlib.colors as mpcolors
import matplotlib.cbook as mpcbook
import pandas as pd
import numpy as np

ann_path = "/path/to/PC_Cap_3D_Dataset/pc_annotation.csv"
ann_csv = pd.read_csv(ann_path)

ann_title_dict: dict[str, int] = {}
for i, title in enumerate(ann_csv):
  ann_title_dict[title] = i

cap_cls_data = [[], []]
for i in range(len(ann_csv)):
  if str.strip(ann_csv.loc[i, "cap_type"]) == "total":
    cap_cls_data[0].append(ann_csv.loc[i, "cap"])
  elif str.strip(ann_csv.loc[i, "cap_type"]) == "coupling":
    cap_cls_data[1].append(ann_csv.loc[i, "cap"])
  else:
    assert "error"
cap_cls_data = np.array(cap_cls_data, dtype=object)

bins = []
start_bins = 1e-22
magnitude_order = 8
for i in range(magnitude_order):
  bins.extend(np.linspace(start_bins * pow(10, i), start_bins * pow(10, i + 1), 10, endpoint=False))
print(bins)

def pie_autopct(pct: float, show_ratio: float) -> str:
  if pct > show_ratio:
    return '{:.2f}%'.format(pct)
  else:
    return ''

def build_labels(frequency: list[float], bins: list[float], show_ratio: float) -> list[str]:
  assert len(frequency) + 1 == len(bins), "frequency and bins mismatch"
  assert 0 < show_ratio < 1, "ratio overflow"
  labels = []
  total_frequency = sum(frequency)
  for i in range(len(frequency)):
    if frequency[i] / total_frequency < show_ratio:
      labels.append("")
    else:
      labels.append("{:.2g}~{:.2g}".format(bins[i], bins[i + 1]))
  return labels


def linear_gradient_color(name: str, start_color: str, end_color: str, num_steps: int) -> list:
    """
    Param
      name: of the output color list
      start_color: the starting color string of list
      end_color: the end color string of list
      num_steps: the interpolated count
    """
    cmap = mpcolors.LinearSegmentedColormap.from_list(
        name, [start_color, end_color], N=num_steps)
    gradient_colors = [cmap(i) for i in range(num_steps)]
    return gradient_colors

def render_patches(name: str, start_color: str, end_color: str, numsteps: int, patches: mpcbook.silent_list) -> None:
  color_map = mpcolors.LinearSegmentedColormap.from_list(name, [start_color, end_color], N=numsteps)
  for id in range(len(patches)):
    plt.setp(patches[id], 'facecolor', color_map(id))

fig, axes = plt.subplots(3, 2, figsize=(12, 16))

colors0_0 = ['#1F77B4', '#FF7F0E']
axe0_0: plt.Axes = axes[0, 0]
axe0_0.hist(cap_cls_data, label=['total', 'coupling'], bins=bins, stacked=True, color=colors0_0)
axe0_0.legend()
axe0_0.set_xscale('log')
axe0_0.set_xlabel('capacitance/F')
axe0_0.set_ylabel('counts')

axe0_1: plt.Axes = axes[0, 1]
axe0_1.pie([len(cap_cls_data[0]), len(cap_cls_data[1])], autopct='%.2f%%')

color_start_end_total = ['#E0FFFF', '#1F77B4']
axe1_0: plt.Axes = axes[1, 0]
counts1_0, bins1_0, patches1_0 = axe1_0.hist(cap_cls_data[0], bins=bins[40 : 60])
render_patches("total", color_start_end_total[0], color_start_end_total[1], len(counts1_0), patches1_0)
axe1_0.set_xscale('log')
axe1_0.set_xlabel('total capacitance/F')
axe1_0.set_ylabel('counts')

colors = linear_gradient_color('total', color_start_end_total[0], color_start_end_total[1], len(counts1_0))
show_ratio_total = 0.1
labels = build_labels(counts1_0, bins1_0, show_ratio_total)
pie_autopct_total = lambda x: pie_autopct(x, 100 * show_ratio_total)
axe1_1: plt.Axes = axes[1, 1]
axe1_1.pie(counts1_0, autopct=pie_autopct_total, labels=labels, colors=colors)
axe1_1.set_title(f"total capacitance ratio > {show_ratio_total}")

color_start_end_coupling = ['#FFD39B', '#CD661D']
axe2_0: plt.Axes = axes[2, 0]
counts2_0, bins2_0, patches2_0 = axe2_0.hist(cap_cls_data[1], bins=bins[10 : 60])
render_patches("coupling", color_start_end_coupling[0], color_start_end_coupling[1], len(counts2_0), patches2_0)
axe2_0.set_xscale('log')
axe2_0.set_xlabel('coupling capacitance/F')
axe2_0.set_ylabel('counts')

colors = linear_gradient_color('coupling', color_start_end_coupling[0], color_start_end_coupling[1], len(counts2_0))
show_ratio_coupling = 0.04
labels = build_labels(counts2_0, bins2_0, show_ratio_coupling)
pie_autopct_coupling = lambda x: pie_autopct(x, 100 * show_ratio_coupling)
axe2_1: plt.Axes = axes[2, 1]
axe2_1.pie(counts2_0, autopct=pie_autopct_coupling, labels=labels, colors=colors)
axe2_1.set_title(f"coupling capacitance ratio > {show_ratio_coupling}")

plt.subplots_adjust(hspace=0.3, wspace=0.2)
plt.show()
```

```python
# samples distribution vs conductor counts in window
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

ann_path = "/path/to/PC_Cap_3D_Dataset/pc_annotation.csv"
ann_csv = pd.read_csv(ann_path, skipinitialspace=True)

ann_title_dict: dict[str, int] = {}
for i, title in enumerate(ann_csv):
  ann_title_dict[title] = i

def query_cdt_num(src: pd.DataFrame, cdt_num: int) -> pd.Series:
  return src["conductor_num"] == cdt_num

def query_total(src: pd.DataFrame) -> pd.Series:
  return src["cap_type"] == "total"

def linear_gradient_color(name: str, start_color: str, end_color: str, num_steps: int) -> list:
    """
    Param
      name: of the output color list
      start_color: the starting color string of list
      end_color: the end color string of list
      num_steps: the interpolated count
    """
    cmap = mpcolors.LinearSegmentedColormap.from_list(
        name, [start_color, end_color], N=num_steps)
    gradient_colors = [cmap(i) for i in range(num_steps)]
    return gradient_colors

cdt_num_max = ann_csv["conductor_num"].max()
total_list = []
coupling_list = []
cdt_nums = []
for cdt_num in range(2, cdt_num_max + 1):
  class_data = ann_csv.loc[query_cdt_num(ann_csv, cdt_num)]
  total_df = class_data.loc[query_total]
  coupling_df = pd.concat([class_data, total_df]).drop_duplicates(keep=False)
  samples_num = total_df.shape[0] / cdt_num
  total_list.append(total_df.shape[0])
  coupling_list.append(coupling_df.shape[0])
  print(cdt_num, total_df.shape[0], coupling_df.shape[0])
  cdt_nums.append(str(cdt_num))

plt.barh(cdt_nums, -np.array(total_list), label="total")
plt.barh(cdt_nums, coupling_list, label="coupling")
# original
x_min, x_max = plt.xlim()
locs, _ = plt.xticks()
# custom
x_label = [ abs(int(num)) for num in locs]
plt.xticks(ticks=locs, labels=x_label)
plt.xlim(x_min, x_max)
plt.title("conductor counts")
plt.legend()
plt.show()
```

```python
# draw test dataset accuracy of model
import os
import sys
import time
import torch
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import tqdm

from torch.utils.data import DataLoader

prj_path = os.path.abspath(os.getcwd() + '/../')
sys.path.append(prj_path)

"""DATA"""
def target_transform(target: torch.Tensor) -> torch.Tensor:
  return target * 1e17

from log.PCT_Cap_6_20240616124642.PcCapDataset import PcCapDataset
ann_path = "/path/to/PC_Cap_3D_Dataset/pc_annotation_coupling_test_clean.csv"
ann_csv = pd.read_csv(ann_path, skipinitialspace=True)
dataset  = PcCapDataset(ann_path, npoints=1024, target_transform=target_transform)
BATCH_SIZE = 64
loader = DataLoader(dataset, batch_size=BATCH_SIZE, num_workers=8)

"""MODEL"""
from log.PCT_Cap_6_20240616124642.PCT_Cap import PCT_Cap
model = PCT_Cap(8, 6, 1024)
checkpoint_path = prj_path + '/checkpoint/PCT-Cap-6_20240616124642E526.pt'
try:
  checkpoint = torch.load(checkpoint_path)  
  model.load_state_dict(checkpoint['model_state_dict'])
except:
  assert "fail to load checkpoint"
model.eval()

from utils import try_gpu
device = try_gpu(-1)
model.to(device)

acc = torch.Tensor()
caps = torch.Tensor()
pred: torch.Tensor = None
target: torch.Tensor = None
sum_time = 0
start_time = time.time()
with torch.no_grad():
  for batch_id, (points, target) in tqdm(enumerate(loader), total = len(loader)):
    a = time.process_time()
    pred = model(points)
    sum_time += time.process_time() - a
    pred = pred.view(len(target))
    round_acc = (pred - target) / target
    acc = torch.cat([acc, round_acc])
    target *= 1e-17
    caps = torch.cat([caps, target])

end_time = time.time()

print(f"dataset size: {ann_csv.shape[0]}, path: {ann_path}")
print(f"model path: {checkpoint_path}")

params_num = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Params num: {params_num}")

temp_path = "../temp.pt"
torch.save(model.state_dict(), temp_path)
file_size = os.path.getsize(temp_path)
print(f"Model size: {file_size}")
os.remove(temp_path)

print(f"prediction runtime: {end_time - start_time} seconds")
print(f"prediction cpu time: {sum_time} seconds")
solver_time = ann_csv["solver_time"].sum()
print(f"solver time: {solver_time} seconds")

total = len(loader.dataset)
acc_5 = torch.sum((torch.abs(acc) < 0.05))
acc_10 = torch.sum(torch.abs(acc) >= 0.10) 
acc_5_10 = total - acc_5 - acc_10
print(f"Model Acc(<5%, 5%~10%, >=10%): {acc_5 / total}, {acc_5_10 / total}, {acc_10 / total}")

abs_acc = torch.abs(acc)
avg_acc = torch.mean(abs_acc)
max_acc = torch.max(abs_acc)
std = torch.std(abs_acc)
print(f"average: {avg_acc}, max: {max_acc}, std: {std}")

fig, axes = plt.subplots(1, 1, figsize=(6, 5))

axe0_0: plt.Axes = axes
axe0_0.scatter(x=acc.detach().numpy(), y=caps.detach().numpy(), s=3)
locs = axe0_0.get_xticks()
x_labels = [ int(100 * num) for num in locs]
axe0_0.set_xticks(ticks=locs, labels=x_labels)
axe0_0.set_xlabel('Relative error (%)')
axe0_0.set_yscale('log')
axe0_0.set_ylabel('Capacitance (F)')

plt.show()
```

```python
# dataset capacitance distribution detail
import matplotlib.pyplot as plt
import matplotlib.colors as mpcolors
import matplotlib.cbook as mpcbook
import pandas as pd
import numpy as np

total_train_clean_path = "/path/to/PC_Cap_3D_Dataset/pc_annotation_total_train_clean.csv"
total_test_clean_path  = "/path/to/PC_Cap_3D_Dataset/pc_annotation_total_test_clean.csv"
coupl_train_clean_path = "/path/to/PC_Cap_3D_Dataset/pc_annotation_coupling_train_clean.csv"
coupl_test_clean_path  = "/path/to/PC_Cap_3D_Dataset/pc_annotation_coupling_test_clean.csv"

total_train_clean_csv = pd.read_csv(total_train_clean_path)
total_test_clean_csv = pd.read_csv(total_test_clean_path)
coupl_train_clean_csv = pd.read_csv(coupl_train_clean_path)
coupl_test_clean_csv = pd.read_csv(coupl_test_clean_path)
total_csv = pd.concat([total_train_clean_csv, total_test_clean_csv])
coupl_csv = pd.concat([coupl_train_clean_csv, coupl_test_clean_csv])

print(f"total_train_clean_csv.shape = {total_train_clean_csv.shape}")
print(f"total_test_clean_csv.shape = {total_test_clean_csv.shape}")
print(f"coupl_train_clean_csv.shape = {coupl_train_clean_csv.shape}")
print(f"coupl_test_clean_csv.shape = {coupl_test_clean_csv.shape}")
print(f"total_csv.shape = {total_csv.shape}")
print(f"coupl_csv.shape = {coupl_csv.shape}")

def linear_gradient_color(name: str, start_color: str, end_color: str, num_steps: int) -> list:
    """
    Param
      name: of the output color list
      start_color: the starting color string of list
      end_color: the end color string of list
      num_steps: the interpolated count
    """
    cmap = mpcolors.LinearSegmentedColormap.from_list(
        name, [start_color, end_color], N=num_steps)
    gradient_colors = [cmap(i) for i in range(num_steps)]
    return gradient_colors

def render_patches(name: str, start_color: str, end_color: str, numsteps: int, patches: mpcbook.silent_list) -> None:
  color_map = mpcolors.LinearSegmentedColormap.from_list(name, [start_color, end_color], N=numsteps)
  for id in range(len(patches)):
    plt.setp(patches[id], 'facecolor', color_map(id))

def pie_autopct(pct: float, show_ratio: float) -> str:
  if pct > show_ratio:
    return '{:.2f}%'.format(pct)
  else:
    return ''

def build_labels(frequency: list[float], bins: list[float], show_ratio: float) -> list[str]:
  assert len(frequency) + 1 == len(bins), "frequency and bins mismatch"
  assert 0 < show_ratio < 1, "ratio overflow"
  labels = []
  total_frequency = sum(frequency)
  for i in range(len(frequency)):
    if frequency[i] / total_frequency < show_ratio:
      labels.append("")
    else:
      labels.append("{:.2g}~{:.2g}".format(bins[i], bins[i + 1]))
  return labels

def plot_total(axes: plt.Axes):
  axe0_0: plt.Axes = axes
  bins = np.linspace(9.8e-18, 1.1e-16, 30)
  counts0_0, bins0_0, patches0_0 = axe0_0.hist(total_csv["cap"], bins=bins)
  color_start_total = '#94D3DC'
  color_end_total = '#1D5681'
  render_patches('total', color_start_total, color_end_total, len(counts0_0), patches0_0)
  axe0_0.set_xscale('log')
  axe0_0.set_xlabel('Capacitance (F)')
  axe0_0.set_ylabel('Count')

def plot_coupling(axes: plt.Axes):
  axe1_0: plt.Axes = axes
  bins = np.linspace(9e-18, 6e-17, 30)
  counts1_0, bins1_0, patches1_0 = axe1_0.hist(coupl_csv["cap"], bins=bins)
  color_start_coupling = '#E8B690'
  color_end_coupling = '#D57832'
  render_patches('coupling', color_start_coupling, color_end_coupling, len(counts1_0), patches1_0)
  # print(counts1_0)
  axe1_0.set_xscale('log')
  axe1_0.set_xlabel('Capacitance (F)')
  axe1_0.set_ylabel('Count')

fig, axes = plt.subplots(1, 1, figsize=(8, 6))
plot_total(axes)
# plot_coupling(axes)
plt.show()
```

```python
# 3D grid sampling
import os, sys
prj_path = os.path.abspath(os.getcwd() + '/../')
sys.path.append(prj_path)

import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as mcm
import matplotlib.transforms as mtrans
import numpy as np
import pandas as pd
import random

ann_path = '/path/to/PC_Cap_3D_Dataset/pc_annotation.csv'
ann = pd.read_csv(ann_path)
idx = random.randint(0, ann.shape[0] - 1)
# idx=61411 # total
idx = 99911 # coup
print(f"idx = {idx}")
data_path: str = os.path.dirname(ann_path) + '/' + ann.loc[idx, "point_cloud_path"]
print(data_path)
data: np.ndarray = np.loadtxt(data_path, delimiter=',').astype(np.float32)

from data_utils.grid_sampler import GridSampler
size = (15,15,6)
sampler = GridSampler(size=size, ann=ann, decay=0.9)
extend_info = ann.loc[idx]
x_interval = np.linspace(extend_info["v1_x"], extend_info["v1_x"] + extend_info["dx"], size[0])
y_interval = np.linspace(extend_info["v1_y"], extend_info["v1_y"] + extend_info["dy"], size[1])
z_interval = np.linspace(extend_info["v1_z"], extend_info["v1_z"] + extend_info["dz"], size[2])

point_num = data.shape[0]
sampling_target = np.zeros((len(data), 4))
grid_d3 = np.zeros(size +(3,))
for point_id in range(point_num):
  p = data[point_id]
  x_idx = sampler.binary_search(x_interval, p[0])
  y_idx = sampler.binary_search(y_interval, p[1])
  z_idx = sampler.binary_search(z_interval, p[2])
  sampling_target[point_id, :3] = (point_id, sampler.pack_grid_id(x_idx, y_idx, z_idx), random.random())
  if p[8] > 0:
    grid_d3[x_idx, y_idx, z_idx, 0] += 1 # main conductor
  elif p[8] < 0:
    grid_d3[x_idx, y_idx, z_idx, 1] += 1 # target environment conductor
  else:
    grid_d3[x_idx, y_idx, z_idx, 2] += 1 # non-target environment conductor

density_grid = sampler.density_strategy(grid_d3, sampler.init_density, sampler.decay)

for point_id in range(point_num):
  x_idx, y_idx, z_idx = sampler.unpack_grid_id(sampling_target[point_id, 1])
  sampling_target[point_id, 3] = sampling_target[point_id, 2] * density_grid[x_idx, y_idx, z_idx]
sampling_target = sampling_target[np.argsort(sampling_target[:, 3])[::-1]]
output_index = sampling_target[:, 0].astype(int)
output_data = data[output_index[: 1024], :]

c_list = ['#FFFFFF', '#F57C0B']
color_norm = mcolors.Normalize(vmin=0, vmax=1)
color_map = mcolors.LinearSegmentedColormap.from_list('density_cm', c_list)
density_grid_cm = color_map(color_norm(density_grid))
voxelarray = np.sum(grid_d3, axis=3) != 0
density_grid_cm[:, :, :, 3] = 0.7

fig = plt.figure(figsize=(12, 12))

axe_0: plt.Axes = fig.add_subplot(221, projection='3d')
axe_0.scatter(data[:, 0], data[:, 1], data[:, 2], s=2)
axe_0.set_xlim(xmin=extend_info["v1_x"], xmax=extend_info["v1_x"] + extend_info["dx"])
axe_0.set_ylim(ymin=extend_info["v1_y"], ymax=extend_info["v1_y"] + extend_info["dy"])
axe_0.set_zlim(zmin=extend_info["v1_z"], zmax=extend_info["v1_z"] + extend_info["dz"])

axe_1: plt.Axes = fig.add_subplot(222, projection='3d')
axe_1.scatter(output_data[:, 0], output_data[:, 1], output_data[:, 2], s=2)
axe_1.set_xlim(xmin=extend_info["v1_x"], xmax=extend_info["v1_x"] + extend_info["dx"])
axe_1.set_ylim(ymin=extend_info["v1_y"], ymax=extend_info["v1_y"] + extend_info["dy"])
axe_1.set_zlim(zmin=extend_info["v1_z"], zmax=extend_info["v1_z"] + extend_info["dz"])

axe_2: plt.Axes = fig.add_subplot(223, projection='3d')
axe_2.voxels(voxelarray, facecolors=density_grid_cm)
axe_2_pos: mtrans.Bbox = axe_2.get_position()
cax_pos = mtrans.Bbox.from_extents(
  axe_2_pos.x1 + 0.02,
  axe_2_pos.y0,
  axe_2_pos.x1 + 0.02 + 0.02,
  axe_2_pos.y1 
)
cax = axe_2.figure.add_axes(cax_pos)
fig.colorbar(mappable=mcm.ScalarMappable(norm=color_norm, cmap=color_map), cax=cax, ax=axe_2, orientation='vertical', label='density')

plt.show()
```
